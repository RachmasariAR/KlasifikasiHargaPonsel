```python
# Split data menjadi data latih dan uji
train_data, test_data = data.randomSplit([0.8, 0.2], seed=42)
```

```python

# Fungsi untuk mengevaluasi model
def evaluate_model(model, test_data):
    predictions = model.transform(test_data)
    evaluator_acc = MulticlassClassificationEvaluator(labelCol="Price_Range", predictionCol="prediction", metricName="accuracy")
    evaluator_precision = MulticlassClassificationEvaluator(labelCol="Price_Range", predictionCol="prediction", metricName="weightedPrecision")
    evaluator_recall = MulticlassClassificationEvaluator(labelCol="Price_Range", predictionCol="prediction", metricName="weightedRecall")
    evaluator_f1 = MulticlassClassificationEvaluator(labelCol="Price_Range", predictionCol="prediction", metricName="f1")

    accuracy = evaluator_acc.evaluate(predictions)
    precision = evaluator_precision.evaluate(predictions)
    recall = evaluator_recall.evaluate(predictions)
    f1 = evaluator_f1.evaluate(predictions)

    return accuracy, precision, recall, f1
```

# 3. Model Gradient Boosting

```python
from pyspark.sql.functions import col, when
from pyspark.ml.classification import GBTClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
import matplotlib.pyplot as plt
import seaborn as sns

# Memastikan kolom 'Price_Range' hanya berisi nilai 0 dan 1
train_data = train_data.withColumn("Price_Range", when(col("Price_Range") == 3.0, 1).otherwise(col("Price_Range")))
test_data = test_data.withColumn("Price_Range", when(col("Price_Range") == 3.0, 1).otherwise(col("Price_Range")))

# Menghapus baris dengan nilai selain 0 atau 1 pada kolom 'Price_Range'
train_data = train_data.filter((col("Price_Range") == 0) | (col("Price_Range") == 1))
test_data = test_data.filter((col("Price_Range") == 0) | (col("Price_Range") == 1))

# Ubah tipe data kolom 'Price_Range' menjadi Double
train_data = train_data.withColumn("Price_Range", col("Price_Range").cast("double"))
test_data = test_data.withColumn("Price_Range", col("Price_Range").cast("double"))

# Inisialisasi GBTClassifier
gbt = GBTClassifier(labelCol="Price_Range", featuresCol="features", maxIter=100, seed=42)

# Melatih model dengan data latih
gbt_model = gbt.fit(train_data)

# Memprediksi pada data uji
gbt_predictions = gbt_model.transform(test_data)

# Evaluasi model
evaluator = MulticlassClassificationEvaluator(labelCol="Price_Range", predictionCol="prediction")

accuracy = evaluator.evaluate(gbt_predictions, {evaluator.metricName: "accuracy"})
precision = evaluator.evaluate(gbt_predictions, {evaluator.metricName: "weightedPrecision"})
recall = evaluator.evaluate(gbt_predictions, {evaluator.metricName: "weightedRecall"})
f1_score = evaluator.evaluate(gbt_predictions, {evaluator.metricName: "f1"})

print("\nHasil Evaluasi Model Gradient Boosting:")
print(f"Akurasi: {accuracy:.4f}")
print(f"Presisi: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")

# Ekstrak nilai prediksi dan label sebenarnya
predictions_and_labels = gbt_predictions.select("prediction", "Price_Range").rdd.map(lambda row: (row[0], row[1]))

# Buat confusion matrix
metrics = MulticlassMetrics(predictions_and_labels)
confusion_matrix = metrics.confusionMatrix().toArray()

# Plot confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(confusion_matrix, annot=True, fmt=".0f", cmap="Blues", xticklabels=["Predicted 0", "Predicted 1"], yticklabels=["True 0", "True 1"])
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix - Gradient Boosting")
plt.show()
```

# 4. Model K-Nearest Neighbors (KNN)

```python
from pyspark.ml.feature import VectorAssembler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Memeriksa dan menghapus kolom 'features' jika sudah ada
if 'features' in train_data.columns:
    train_data = train_data.drop('features')
if 'features' in test_data.columns:
    test_data = test_data.drop('features')

# Persiapkan data dan fitur
assembler = VectorAssembler(inputCols=[col for col in train_data.columns if col != 'Price_Range'], outputCol="features")
train_data = assembler.transform(train_data)
test_data = assembler.transform(test_data)

# Memeriksa dan menangani nilai yang hilang
train_data = train_data.dropna()
test_data = test_data.dropna()

# Model KNN - kita implementasikan perhitungan jarak secara manual
def knn_predict(test_point, train_points, k=5):
    distances = []
    for row in train_points:
        dist = np.linalg.norm(test_point - row)
        distances.append((dist, row))
    distances.sort(key=lambda x: x[0])
    nearest_neighbors = distances[:k]

    # Menghindari nilai NaN saat mengambil label
    if len(nearest_neighbors) > 0:
        prediction = np.mean([neighbor[1] for neighbor in nearest_neighbors])
        return prediction
    else:
        return 0  # Mengembalikan 0 jika tidak ada tetangga yang valid

# Prediksi dengan KNN secara manual
predictions = []
for test_row in test_data.select("features").collect():
    test_point = np.array(test_row.features)
    predictions.append(knn_predict(test_point, train_data.select("features").collect(), k=5))

# Pastikan prediksi tidak mengandung nilai NaN
predictions = [pred if not np.isnan(pred) else 0 for pred in predictions]

# Evaluasi hasil
y_test = test_data.select("Price_Range").collect()
y_pred_knn = predictions

# Menghitung akurasi, precision, recall, f1 score
knn_accuracy = accuracy_score(y_test, y_pred_knn)
knn_classification_report = classification_report(y_test, y_pred_knn)
knn_confusion_matrix = confusion_matrix(y_test, y_pred_knn)

# Menampilkan hasil
print("\nK-Nearest Neighbors Results:")
print(f"Accuracy: {knn_accuracy:.4f}")
print(knn_classification_report)

# Plot confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(knn_confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['True Negative', 'True Positive'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix - K-Nearest Neighbors')
plt.show()
```

# Hyperparameter Tuning

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.classification import RandomForestClassifier, GBTClassifier, NaiveBayes
import numpy as np
from sklearn.metrics import accuracy_score
from pyspark.sql.functions import when, col

# Inisialisasi Spark Session
spark = SparkSession.builder.appName("PySpark Models").getOrCreate()

# Load dataset
df = spark.read.csv("train.csv", header=True, inferSchema=True)

# Ubah label 'Price_Range' menjadi tipe integer untuk klasifikasi multi-class
df = df.withColumn("Price_Range", when(col("Price_Range") == 3, 1).otherwise(0))

# Fitur dan label
feature_columns = [col for col in df.columns if col != "Price_Range"]
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df = assembler.transform(df).select("features", "Price_Range")

# Split data menjadi training dan testing
train, test = df.randomSplit([0.7, 0.3], seed=42)

# Model 1: Random Forest
rf = RandomForestClassifier(labelCol="Price_Range", featuresCol="features", numTrees=100, seed=42)
rf_model = rf.fit(train)
rf_predictions = rf_model.transform(test)

# Evaluasi Random Forest
rf_evaluator = MulticlassClassificationEvaluator(labelCol="Price_Range", predictionCol="prediction", metricName="accuracy")
rf_accuracy = rf_evaluator.evaluate(rf_predictions)
print(f"Random Forest Accuracy: {rf_accuracy:.2%}")

# Model 2: Naive Bayes
nb = NaiveBayes(labelCol="Price_Range", featuresCol="features", modelType="multinomial")
nb_model = nb.fit(train)
nb_predictions = nb_model.transform(test)

# Evaluasi Naive Bayes
nb_evaluator = MulticlassClassificationEvaluator(labelCol="Price_Range", predictionCol="prediction", metricName="accuracy")
nb_accuracy = nb_evaluator.evaluate(nb_predictions)
print(f"Naive Bayes Accuracy: {nb_accuracy:.2%}")

# Model 3: Gradient Boosting (multi-class classification)
gb = GBTClassifier(labelCol="Price_Range", featuresCol="features", maxIter=100, seed=42)
gb_model = gb.fit(train)
gb_predictions = gb_model.transform(test)

# Evaluasi Gradient Boosting
gb_evaluator = MulticlassClassificationEvaluator(labelCol="Price_Range", predictionCol="prediction", metricName="accuracy")
gb_accuracy = gb_evaluator.evaluate(gb_predictions)
print(f"Gradient Boosting Accuracy: {gb_accuracy:.2%}")

# Model 4: K-Nearest Neighbors (KNN)
# Konversi dataset ke NumPy
train_features = np.array(train.select("features").rdd.map(lambda x: x[0].toArray()).collect())
train_labels = np.array(train.select("Price_Range").rdd.map(lambda x: x[0]).collect())
test_features = np.array(test.select("features").rdd.map(lambda x: x[0].toArray()).collect())
test_labels = np.array(test.select("Price_Range").rdd.map(lambda x: x[0]).collect())

# Fungsi untuk prediksi KNN
def knn_predict(test_point, train_features, train_labels, k=5):
    distances = np.linalg.norm(train_features - test_point, axis=1)
    nearest_indices = np.argsort(distances)[:k]
    nearest_labels = train_labels[nearest_indices]
    return np.argmax(np.bincount(nearest_labels.astype(int)))

# Prediksi untuk test set
knn_predictions = [knn_predict(test_point, train_features, train_labels, k=5) for test_point in test_features]

# Evaluasi KNN
knn_accuracy = accuracy_score(test_labels, knn_predictions)
print(f"KNN Accuracy: {knn_accuracy:.2%}")

# Hasil Akhir: Bandingkan Semua Model
all_results = [
    (rf_accuracy, "Random Forest"),
    (nb_accuracy, "Naive Bayes"),
    (gb_accuracy, "Gradient Boosting"),
    (knn_accuracy, "KNN"),
]

# Pilih 2 model terbaik berdasarkan akurasi
best_two_models = sorted(all_results, key=lambda x: x[0], reverse=True)[:2]
print("\nBest Two Models:")
for model in best_two_models:
    print(f"{model[1]} with Accuracy: {model[0]:.2%}")
```
